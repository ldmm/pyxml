```python
f = open('class_materials/1changin01.law.txt','r')

counter = 0
allLines = []

#reading
for line in f:
    if len(line.strip()):
        counter += 1
        print(str(counter) + '\t'+ 'line' + str(counter) + '\t' + line.strip())
        allLines += [str(counter) + '\t'+ 'line' + str(counter) + '\t' + line.strip()]
        
f.close()

#writing    
f = open('output.txt', 'w')
for item in allLines:
    f.write("%s\n" % item)
f.close()
    
```
