```python
from lxml import etree
import datetime

from bs4 import BeautifulSoup
import lxml

import re

#because we don't have any markup, we will use etree here; 
#beautiful soup +lxml is for when you have that stuff already

allLines = []
rootElement = etree.Element("paradise")


fo = open('../class_materials/paradise.lost.1.txt', 'r')

for l in fo.readlines():
    allLines += [l.strip()]    
        
fo.close()


lineNumber = 0
stanzaNumber = 0
stanzaElement = etree.SubElement(rootElement, 'stanza')
characterList = ["Father", "Jesus", "Gabriel", "Raphael", "Uriel", "Abdiel", "Michael", "Satan", "Beelzebub", "Belial", "Mammon", "Mulciber", "Moloch", "Sin", "Death", "Adam", "Eve"]

for l in allLines:
   # l.replace(\_+a-z+A-Z\_, "") #trying to take the underscores out
    if l:
        if len(l) > 10: 
            lineNumber +=1
            lineElement = etree.SubElement(stanzaElement, 'line')
            lineElement.set('n', str(lineNumber))
            
            listOfWords = l.split()
            
            tailFlag = False
            tailFlag2 = False
            
            for w in listOfWords:
                     
               # print(w)
                
                if w[0] == "_" or w[-1]=="_":
                    hiElement = etree.SubElement(lineElement, 'hi')
                    hiElement.text = w
                    tailFlag=True                    
                
                else:
                    if tailFlag:
                        hiElement.tail = w
                        tailFlag = False
                    
                    elif tailFlag2:
                        for p in l:
                            p = p.strip()
                            if p in characterList:
                                nameElement = etree.SubElement(lineElement, 'persName')
                                #nameElement.set('type', 'persName')
                                nameElement.text = p
                            
                    else:
                        if lineElement.text:
                            lineElement.text += w + ' '
                        else:
                            lineElement.text = w + ' '
               
    else:
        stanzaNumber +=1
        stanzaElement = etree.SubElement(rootElement, 'stanza')
        stanzaElement.set('type', 'stanza')
        stanzaElement.set('n', str(stanzaNumber))
        
            
xmlString = etree.tostring(rootElement, pretty_print=True, encoding ='UTF-8')
fh = open('../output/paradise.xml', 'wb') #change the output folder as needed
fh.write(xmlString)
fh.close()


print('done')
```
