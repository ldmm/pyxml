```python
#1- import statements
from bs4 import BeautifulSoup
import lxml
```

```python
#2- open file
f = open('class_materials/dap/hamlet.xml','r')
content = f.read()
f.close()
```

```python
#3- create soup
soup = BeautifulSoup(content, 'lxml')
```

```python
#4- parse
for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
```

```python
#find all the scenes in each act
for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
    
    for scene in act.find_all('div',type='scene'):
        print('scene', scene['n'])
```

```python
#print all the scenes that have even numbers
for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
    
    for scene in act.find_all('div',type='scene'):
        if int(scene['n']) % 2 ==0:
            print('scene', scene['n'])
```

```python
#print all the characters in all the scenes
for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
    
    for scene in act.find_all('div',type='scene'):
        print('scene', scene['n'])
        
        for speaker in scene.find_all('sp'):
            print('speaker', speaker['who'])
        
        
```

```python
#print all the acts and scene numbers where #First_Clown is character
for act in soup.find_all('div',type='act'):
    
    for scene in act.find_all('div',type='scene'):
        
        for speaker in scene.find_all('sp'):
            
            if speaker['who'] == '#First_Clown':
                print(act['n'], scene['n'])
```

```python
#print all the acts and scene numbers where #First_Clown is character
#same, but a different approach

actsAndScenes = []

for speaker in soup.find_all('sp',who='#First_Clown'):
    #print(speaker.parent.parent['n'], speaker.parent['n'] )
    
    temp = speaker.parent.parent['n'] + '.' + speaker.parent['n']
    #if temp not in actsAndScenes:
    actsAndScenes.append(temp)

print(set(actsAndScenes))
```

```python
#print all the lines from #First_Clown
for speaker in soup.find_all('sp',who='#First_Clown'):
    for c in speaker.children:
        if c.name:
            print(c.text)
```
