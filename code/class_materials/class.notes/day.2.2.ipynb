```python
#1- import statements
from bs4 import BeautifulSoup
import lxml

#2- open file
f = open('class_materials/dap/comedyerrors.xml','r')
content = f.read()
f.close()

#3- create soup
soup = BeautifulSoup(content, 'lxml')
```

```python
#4- parse
for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
```

```python
#print all the acts and scenes
for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
    
    for scene in act.find_all('div',type='scene'):
        print('scene', scene['n'])
        
    print('*'*30)
```

```python
#print all the characters in every act and scene
for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
    
    for scene in act.find_all('div',type='scene'):
        print('scene', scene['n'])
        
        for speaker in scene.find_all('sp'):
            print('speaker:', speaker['who'][1:])
        
        print('*'*10)
        
    print('*'*30)
    
    
```

```python
#print all lines by Aegeon in every act and scene

for act in soup.find_all('div',type='act'):
    print('act:', act['n'])
    
    for scene in act.find_all('div',type='scene'):
        print('scene', scene['n'])
        
        for speaker in scene.find_all('sp',who='#Aegeon'):
            print('speaker:', speaker['who'])
            print(speaker.text)
        
        print('*'*10)
        
    print('*'*30)
```
